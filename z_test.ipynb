```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.graphics.gofplots as smg
```

```python
n = 1000
sample = np.random.normal(loc=100, scale=30, size=n)
print(np.mean(sample), np.std(sample))
```

```python
stat, p = stats.shapiro(sample)

print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Принимаем нулевую гипотезу: распределение нормальное')
else:
    print('Отвергаем нулевую гипотезу: распределение не является нормальным')
```

```python
fig, ax = plt.subplots(figsize=(10, 3), nrows=1, ncols=2)
sns.histplot(sample, ax=ax[0])
smg.qqplot(sample, line='s', ax=ax[1])
plt.show()
```

Сравним среднее в выборке и среднее в генеральной совокупности

```python
mean = 103 # среднее в гс
sample_mean = np.mean(sample) 
std = 30 # стандартное отклонение в гс
se = std/np.sqrt(n) # стандартная ошибка

print((mean-sample_mean)/se)
```

```python
from statsmodels.stats.weightstats import ztest as ztest
stat, p_value = ztest(sample, value=103)
print('stat=%.3f, p=%.8f' % (stat, p_value))
```
